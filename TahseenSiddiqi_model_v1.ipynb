```python
#Importing all the desired packages and modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.models import load_model
```

```python
data=pd.read_csv('wdbc.data.csv',header=None)
data.head()
```

```python
data.info()
```

#By reading the feature names we know that column 1 is the target value which has 2 classe Diagnosis
#(M = malignant, B = benign)
#Column 1 is the id column which means nothing for us.

```python
#Checking for unique values of column 1 id and dropping it
data[0].nunique()
```

```python
#Since the first column which is column 0 all are unique values which has id stored we will drop it
data.drop(columns=[0],axis=1,inplace=True)
```

```python
data.head()
```

```python
#Since we don't have any missing values , we don't need preprocessing of the data apart from assigning numbers of M and B.
#We are replacing M with 0 and B with 1
data[1]=data[1].apply(lambda x: 0 if x == 'M' else 1)
```

```python
data[1].value_counts()
```

```python
#Diving X,y and train and test dataset
X=data.drop([1],axis=1)
Y=data[1]
dim=len(X.columns)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 32)
```

```python
#Creating Simple ANN
model = Sequential()
model.add(Dense(dim, input_dim=dim, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=250, batch_size=32)
```

```python
loss = model.evaluate(X_test, y_test, verbose=2)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
predicted=[1 if x > 0.5 else 0 for x in y_pred]
print(predicted)
```

```python
#prediction and evalution
import numpy as np
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predicted)
```
